#### 1)  crie uma série do pandas a partir de uma lista com os dados abaixo:

Em um estudo sobre alteração na temperatura global, A NASA disponibiliza dados de diferenças de temperatura média da superfície terrestre relativos às médias de temperatura entre 1951 e 1980. Os dados originais podem ser vistos no site da NASA/GISS, e estão dispostos a cada década na tabela abaixo.

|ano|anomalia termica|
|:-:|:----:|
| 1900 | -0.08 |
| 1920 | -0.27 |
| 1940 | 0.12 |
| 1960 | -0.03 |
| 1980 | 0.26 |
| 2000 | 0.40 |
| 2020 | 1.02 |

Crie uma séries do Pandas a partir de uma lista com esses dados.

```python
# Importar as bibliotecas necessárias
import pandas as pd
import numpy as np
```

```python
diferenca_temp = [-0.08, -0.27, 0.12, -0.03, 0.26, 0.40, 1.02]
diferencaSeriestemp = pd.Series(data = diferenca_temp)
diferencaSeriestemp
```

#### 4) Transforme o ndarray abaixo em um dataframe. 
O numpy é capaz de gerar arrays n-dimensionais com números pseudo-aleatórios de acordo com uma variedade de distribuições, como no exemplo abaixo. Transforme esse nd-array em um DataFrame.

#### 5) Nomeie os índices das linhas com inteiros de 1 a 20, e as colunas com os nomes "x1", "x2", e "x3" respectivamente.

```python
#seu código aqui
lista_20 = list(range(1, 21)) # criar uma lista de números do 1 ao 20
arraydf.index = lista_20
arraydf.columns = ['x1', 'x2', 'x3']
arraydf
```

#### 6) No DataFrame do exercício 5, crie uma nova coluna como sendo a média das três colunas, e dê a ela o nome de "media" (não recomendo colocar acentos em nomes de variáveis).

```python
# seu código aqui
# Fazer a média aritmética
arraydf['media'] = (arraydf['x1'] + arraydf['x2'] + arraydf['x3']) / 3
arraydf
```
