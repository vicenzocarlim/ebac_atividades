####  2) Melhorando a interação com o API
Observe dois URLs de consultas diferentes, por exemplo o URL utilizado em aula, e o URL feito no exercício anterior. Compare-os e observe as diferenças.

1. Faça uma função em Python que recebe como argumento o UF da consulta e o tipo de interessado (```'Estado'```ou ```Município```), e que devolve os dados da consulta no formato *DataFrame*.
2. Quantas solicitações para o Estado podem ser consultadas para Minas Gerais com *status* em 'Arquivado por decurso de prazo' estão registradas?
3. Qual é o município da Bahia com mais solicitações deferidas?
4. Salve um arquivo .csv com os dados de solicitações da Bahia, com interessado = 'Estado'

````python
# Comparação de URLs distintas

url_aula = 'https://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf=SP&tipo_interessado=Estado'
r_aula = requests.get(url_aula)
d_json_aula = r_aula.json()
df_aula = pd.DataFrame(d_json_aula['items'])
df_aula.head()
````

````python
url_tarefa = 'https://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf=RJ&tipo_interessado=Munic%C3%ADpio'
r_tarefa = requests.get(url_tarefa)
d_json_tarefa = r_tarefa.json()
df_tarefa = pd.DataFrame(d_json_tarefa['items'])
df_tarefa.head()
````

*Comparação de URLs*

Pode-se perceber que as diferenças presentes na estrutura das URLs mencionadas dizem respeito à escolha das variáveis ```tipo_interessado``` e ```uf```, sendo as únicas partes da URL que variam de acordo com o interesse.

````python
# 1)
def constroi_data_frame(uf: str, tipo_interessado: str):
    uf = uf.upper()
    tipo_interessado = tipo_interessado.capitalize()
    if tipo_interessado == 'Município':
        tipo_interessado = 'Munic%C3%ADpio' # de acordo com a URL do API
    
    url = f'https://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf={uf}&tipo_interessado={tipo_interessado}'
    r = requests.get(url)
    data_json = r.json()
    df = pd.DataFrame(data_json['items']) # transformar em DataFrame
    return df
````

````python
# 2)
df_mg = constroi_data_frame('mg', 'estado')
df_mg.head()
````

````python
num1 = df_mg[df_mg['status'] == 'Arquivado por decurso de prazo'].shape[0]
print('Há %d solicitação para o Estado em Minas Gerais com status de "Arquivado por decurso de prazo"' % num1)
````

````python
# 3)
df_ba = constroi_data_frame('ba', 'município')
df_ba.head()
````

````python
df_ba[df_ba['status'] == 'Deferido']['interessado'].value_counts()
````

````python
munic = df_ba[df_ba['status'] == 'Deferido']['interessado'].value_counts().index[0]
print(f'O município da Bahia com mais solicitações deferidas é {munic}.')
````

````python
# 4)
df_ba2 = constroi_data_frame('ba', 'estado')
df_ba2.to_csv('PVL_Bahia_Estado.csv', index = False, sep = ';')
````
